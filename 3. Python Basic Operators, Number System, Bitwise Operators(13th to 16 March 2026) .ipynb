# 5. Left Shift (<<):

```python
# In Left shifts, We gain the bit at the end (i.e., adding 0's)
```

```python
12 << 1     # 12 = 1100  Now we have to add 1 0's at last --> 11000. Now find the value of it.
```

```python
12 << 2     # 12 = 1100  Now we have to add 2 0's at last --> 110000. Now find the value of it.
```

```python
25 << 1       
```

# 6. Right Shift (>>):

```python
# In Right Shifts, we lose the bit at the end (be it 0 or 1).
```

```python
12 >> 1      # 12 = 1100, now we have to remove the last bit----> 110. Now find the value.
```

```python
13 >> 3      # 13 = 00001101, removed last 3 --> 00001.
```

```python
25 >> 2
```
