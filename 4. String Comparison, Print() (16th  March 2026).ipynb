# String Comparison:

```python
# Python compares strings using ASCII (Unicode) values:
```

```python
# Python compares strings using ASCII (Unicode) values character by character

a = "Alice"
b = "Bob"

print(a > b)   # Compares 'A' (65) with 'B' (66) → 65 < 66 → False
print(a < b)   # 65 < 66 → True

# ASCII order:
# A(65) < B(66) < C(67) < ... < Z(90)
# a(97) < b(98) < c(99) < ... < z(122)
# Uppercase letters come before lowercase letters
```

```python
# ord() converts a character into its ASCII (Unicode) value

print(ord('a'))  # 97
print(ord('A'))  # 65
print(ord('z'))  # 122
print(ord('Z'))  # 90
```

```python
name3 = 'priyanaka'
name4 = 'pc'
print(name3 <= name4)  # A(65) < B(66) < C(67) < ... < Z(90)
                       # a(97) < b(98) < c(99) < ... < z(122)
```

# 4. Print with end statement:

```python
print('hello')          # 1st statement 
print("good afternoon") # 2nd statement) 
```

```python
# end - used To Control what Appears after the Printed Text 
#  instead of new line, we use "end".
print('hello', end ='--->')  # 1st statement 
print('good afternoon')      # 2nd statement
```
