# Indexing:

# numpy masking or filter:

```python
mat = np.arange(0,100).reshape(10,10)
mat
```

```python
id(mat)
```

```python
mat.shape
```

```python
mat < 50
```

```python
mat[mat < 50]    # Masking
                 # mat mat writing it two times, it is called masking.
```

```python
mat[mat <= 50]
```

```python
mat[mat==50]
```

```python
mat[mat!=50] # not equal to 50, except 50 all will be printed
```

```python
mat[mat>=50]
```
