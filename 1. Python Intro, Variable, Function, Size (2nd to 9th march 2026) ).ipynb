# Variable:

```python
# Two types of Variables:
# 1. Global Variables:
#     The Variables that are declared outside the functions are called as Global Variable.
#     These type of variables can be accessed or invoked by any function. )
# 2. Local Variables:
#     The Variables that are declared inside the functions are called as Local Variable.
#     These type of variables can be accessed only inside the function )
```

# Size of the Bits: 

```python
value = 10
import sys
sys.getsizeof(value) 
# Constructor is the reason, it stores the data of values(object) in memory in the form of "BYTES"
```

```python
value = 20
import sys
sys.getsizeof(value)
```

```python
value = [10,20,44,75]
import sys
sys.getsizeof(value)   
# Constructor will save all this in memory, and that will keep on increase with each new object.
```
