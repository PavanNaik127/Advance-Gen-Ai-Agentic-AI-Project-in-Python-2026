# Slicing: --> String Slicing

```python
# Slicing:
# 1. a[startposition : endposition]   # startposition : endposition+1
# 2. a[startposition :]
# 3. a[: endposition]
# 4. a[:]
```

```python
s = 'welcome'       # Forward Slicing
s[1:4]              # Range: 1 to 4 # for Range, we use Formula as a[startposition:N+1].
```

```python
s[:]               # Range: full
```

```python
s[:5]             # Range: starting to 5th index
```

```python
s[3:]            # Range: 3rd index to ending
```

```python
s[2:9]           # Range: 2nd index to 9th index. even though we have only 7 letters. 
```

```python
s[-5:-2]         # Range: a[-5:-3+1]
```

```python
s[::1]           # Forward Slicing
# Range: forward counting with 1 alternative.
```

```python
s[::2]               # Range: forward counting with 2 alternative.
```

```python
s[::3]               # Range: forward counting with 3 alternative.
```

```python
s[::4]               # Range: forward counting with 4 alternative.
```

```python
s[::-1]              # Backward Slicing
# Range: backward counting with 1 alternative.
```

```python
s[::-3]               # Range: backward counting with 3 alternative.
```

```python
s[2:9:2]         # This is called "Advance String".
#                     ''' # Range: start = 2
#                                   stop = 9 (but string ends at 6)
#                                   step = 2 ( counting with 2 alternatives )'''
# w e (l) c (o) m (e) --> started with l, then with 2 alternative gaps, take next letters.
```

```python
s[2:9:3]              # Range: start = 2
                      #        stop = 9 (but string ends at 6)
                      #        step = 3 ( counting with 3 alternative )'''
```

```python
s = 'welcome'  
s[::-1]
```

```python
s = 'welcome'         '''# Range: start = 5
                                  stop = 1 (but string ends at 6)
                                  step = -1 (backward counting with 1 alternative )'''

'''
Step -1 means move backwards.  |  Start at index 5 → m  
Move backward until just before index 1.  |  Indexes taken: 5 → 4 → 3 → 2
'''
s[5:1:-1]
```

```python
s[5:1:-2]
```

```python
s[5:2:-1]
```

```python
s[2:5:-1]   
#Python would need to go from 1 backwards to 5, which is impossible because 5 is to the right. 
# So Python finds no characters.
# When step is negative (-1): start index must be greater than stop index
```

```python
s[6:1:-2]
```
