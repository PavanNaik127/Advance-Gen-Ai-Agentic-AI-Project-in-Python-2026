# c. input() with string Slicing:

```python
st = input('Enter a string')[0]
print(st)
```

```python
st = input('Enter a string')[0:2]
print(st)
```

```python
st = input('Enter a string')[9:13]
print(st)
```

```python
text = input("Enter text: ")
print(text[0:3])
```
